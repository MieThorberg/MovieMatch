```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn.cluster import estimate_bandwidth
from sklearn.cluster import MeanShift
```

```python
data = pd.read_csv("../data/movies_metadata.csv")
data.head()
```

```python
# drop irrelevant columns
data.drop(['adult', 'imdb_id', 'budget','id','genres', 'title', 'belongs_to_collection', 'homepage', 'original_title', 'overview','poster_path', 'production_companies', 'production_countries', 'release_date', 'revenue', 'runtime', 'spoken_languages', 'status', 'tagline', 'video', 'vote_average', 'vote_count'], 'columns', inplace=True)
data.head()
```

```python
# convert data

# one-hot encoding of 'original_language'
data = pd.get_dummies(data, columns=['original_language'])
data = data.iloc[:200]
data
```

```python
# Find missing values in the data and drop those row:
print('rows before drop: {}'.format(len(data)))
bool_matrix = data.isnull() # dataframe with True and False values for each cell in the data
only_null_filter = bool_matrix.any(axis=1) # is there a True value in any column in each row. returns a pandas Series with index matching index of data dataframe
missing = data[only_null_filter] # show all rows that has one or more null values


# remove null value rows
data = data.dropna()
print("rows after drop: {}".format(len(data)))
data

pd.options.display.max_rows = None
bool_matrix
```

```python
only_null_filter
```

```python
from sklearn.cluster import estimate_bandwidth
bw = estimate_bandwidth(data)
```

```python
analyzer = MeanShift(bandwidth=bw)
analyzer.fit(data)
```

```python
labels = analyzer.labels_
print(labels)
print('\n\n',np.unique(labels))
```

```python
#We will add a new column in dataset which shows the cluster the data of a particular row belongs to.

# create a new column in the dataset
data['cluster_group'] = np.nan
for i in range(len(data)): # loop 714 rows
    data.iloc[i,data.columns.get_loc('cluster_group')] = labels[i] #set the cluster label on each row

data
```
